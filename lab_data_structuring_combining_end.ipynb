```python

import pandas as pd

#url_1 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
df_1 = pd.read_csv('dataset_lab_cleaned.csv')

url_2 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv'
df_2 = pd.read_csv(url_2)

url_3 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv'
df_3 = pd.read_csv(url_3)
```

```python
df_1 = df_1.drop(columns=['Unnamed: 0'])
#df_2 = df_2.drop(columns=['Unnamed: 0'])
#df_3 = df_3.drop(columns=['Unnamed: 0'])
```

```python
def clean_column_names(df):
    """
    Cleans and standardizes column names in a DataFrame.
    """
    df.columns = [col.lower().replace(' ', '_') for col in df.columns]
    # Rename 'st' to 'state' for consistency
    if 'st' in df.columns:
        df = df.rename(columns={'st': 'state'})
    return df
```

```python
# Use function to standardize columns names

df_2 = clean_column_names(df_2)
df_3 = clean_column_names(df_3)
```

```python
# Handle individual datasets invalid values and formats

# df_2

df_2['customer_lifetime_value'] = df_2['customer_lifetime_value'].str.replace('%','').astype(float)
df_2['number_of_open_complaints'] = df_2['number_of_open_complaints'].str.extract('(\d)').astype(float)
```

```python
#df_3 - scaling 'customer_lifetime_value'
df_3['customer_lifetime_value'] = df_3['customer_lifetime_value'] * 100
df_3.head()
```

```python
# Combine dataset:
#   Ensure column are consistent

df_2_reordered = df_2[df_1.columns]
df_3_reordered = df_3[df_1.columns]
```

```python
# Concatenate the 3 dataframes

full_df = pd.concat([df_1, df_2_reordered, df_3_reordered], ignore_index=True)
full_df
```

```python
# handle null values 

def handle_null_values(df):
    """
    Fills null values in a DataFrame using appropriate strategies.
    """
    # Fill categorical columns with the mode
    for col in ['education', 'gender', 'policy_type', 'vehicle_class']:
        if df[col].isnull().any():
            mode_value = df[col].mode()[0]
            df[col] = df[col].fillna(mode_value)
            
    # Fill numerical columns with the median
    for col in ['customer_lifetime_value', 'income', 'monthly_premium_auto', 'total_claim_amount']:
        if df[col].isnull().any():
            median_value = df[col].median()
            df[col] = df[col].fillna(median_value)
            
    return df
```

```python
full_df = handle_null_values(full_df)
full_df.isnull()
```

```python
# Using .value_counts() method, one can assess that there are
# not duplicate values in the dataframe.

full_df.value_counts()
```

```python
pivot_table_clv
```

```python
top_group = pivot_table_clv['customer_lifetime_value'].idxmax()
top_clv = pivot_table_clv['customer_lifetime_value'].max()
```

```python
lowest_group = pivot_table_clv['customer_lifetime_value'].idxmin()
lowest_clv = pivot_table_clv['customer_lifetime_value'].min()
```

```python
print(f"\nAnalysis:")
print(f"The group with the highest average customer lifetime value is {top_group}, with an average of ${top_clv}.")
print(f"The group with the lowest average customer lifetime value is {lowest_group}, with an average of ${lowest_clv}.")
print("\nWe can also observe that within each education level, there isn't a massive difference between genders, indicating that education level might be a stronger factor in determining CLV.")
```

## Bonus

You work at the customer service department and you want to know which months had the highest number of complaints by policy type category. Create a summary table showing the number of complaints by policy type and month.
Show it in a long format table.

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*

```python
# First: Change the column 'effective_to_date' to date time in order
# to use the data to find the month and year information

df_mkt['effective_to_date'] = pd.to_datetime(df_mkt['effective_to_date'])

df_mkt['month'] = df_mkt['effective_to_date'].dt.month
```

```python
complaints_by_month_policy = df_mkt.groupby(['policy_type', 'month'])['number_of_open_complaints'].sum().reset_index()


complaints_by_month_policy.columns = ['policy_type', 'month', 'total_complaints']

print(complaints_by_month_policy)
```

From the table, the customer service department can easily see the number of complaints for each policy type, broken down by month. 
This is more useful than a total count, as it allows them to identify any seasonal trends or specific months where a certain policy type experiences a spike. 
    
    Example, if 'Personal Auto' complaints rise significantly in month 2 (February), the team could investigate if there were specific events or campaigns during that time that might have contributed to the increase.
